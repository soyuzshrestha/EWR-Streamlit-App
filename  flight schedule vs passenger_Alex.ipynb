```python
lst5 = []
for s in df_result['check_in_time']:
    lst5.append(datetime.strptime(s, '%Y/%m/%d %H:%M'))
```

```python
lst6 = []
for s in df_result['flight_time']:
    lst6.append(datetime.strptime(s, '%Y/%m/%d %H:%M'))
```

```python
#Some visualization
plt.rcParams['figure.figsize'] = [12,8]
ax = plt.subplot()
ax.hist(df_v['hours_diff'], color = 'blue', edgecolor = 'black',bins = 27)

# Title and labels
ax.set_title('Passenger Behavior Check', size = 30)
ax.set_xlabel('Check-in hours before the flight schedule', size = 22)
ax.set_ylabel('Number of Passengers', size= 22)

plt.tight_layout()
plt.show()
```
